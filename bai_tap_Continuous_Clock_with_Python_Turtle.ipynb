```python
import datetime
import turtle

screen=turtle.Screen()
screen.bgcolor('blue')
turtle.tracer (0,0)

class DongHo:
    def __init__ (self, ban_kinh):
        self.ban_kinh=ban_kinh
        self.vach_so = turtle.Turtle ()
        self.kim=turtle.Turtle()
        self.gio_hien_tai()
        self.vach_so.hideturtle()
        self.kim.hideturtle()
        
    def gio_hien_tai(self):
        d=datetime.datetime.now()
        self.gio=d.hour
        self.phut=d.minute
        self.giay=d.second
        
    def chay(self):
        self.ve_vach_so()
        self._chay()
    
    def _chay(self):
        self.gio_hien_tai()
        self.ve_kim()
        turtle.update()
        turtle.ontimer(self._chay, 1000)
        
    def ve_vach_so(self):
        self.vach_so.clear()
        #ve vong tron
        self.vach_so.up()
        self.vach_so.goto(0, -self.ban_kinh)
        self.vach_so.down()
        self.vach_so.fillcolor('white')
        self.vach_so.begin_fill()
        self.vach_so.circle(self.ban_kinh)
        self.vach_so.end_fill()
        self.vach_so.up()
        
        #ve vach so
        self.vach_so.goto(0,0)
        self.vach_so.dot(10)
        
        for goc in range (0, 360, 6):
            self.vach_so.up()
            self.vach_so.goto(0,0)
            self.vach_so.setheading(90 - goc)
            self.vach_so.forward(self.ban_kinh-(self.ban_kinh*0.06))
            self.vach_so.down()
            self.vach_so.forward(self.ban_kinh*0.03)
            
        for goc in range (0, 360, 30):
            self.vach_so.up()
            self.vach_so.goto(0,0)
            self.vach_so.setheading (90 - goc)
            self.vach_so.forward(self.ban_kinh - (self.ban_kinh*0.1))
            self.vach_so.down()
            self.vach_so.forward(self.ban_kinh*0.06)
            
    def ve_kim(self):
        print (self.gio, self.phut, self.giay)
        #kim gio
        self.kim.clear()
        self.kim.up()
        self.kim.goto(0,0)
        self.kim.setheading(
            90 
            - (        
                (((self.gio%12)*3600 + self.phut *60 + self.giay)/(12*3600))
                * 360
            )
        )
        
        self.kim.down()
        self.kim.color('black')
        self.kim.forward(self.ban_kinh*0.5)
        #kim phut
        self.kim.up()
        self.kim.goto(0,0)
        self.kim.setheading(90 - (((self.phut*60 + self.giay)/3600)*360))
        self.kim.down()
        self.kim.color('black')
        self.kim.forward(self.ban_kinh*0.65)
        # kim giay
        self.kim.up()
        self.kim.color('red')
        self.kim.goto (0,0)
        self.kim.dot(5)
        self.kim.setheading(90-((self.giay/60)*360))
        self.kim.down()
        self.kim.forward(self.ban_kinh*0.85)
        
dong_ho=DongHo(200)
dong_ho.chay()

turtle.done()
        
            
```
